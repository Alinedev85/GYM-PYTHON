```python

data3 = {
    'A': [1, 2, None, 4, 5],
    'B': [10, None, 30, 40, 50]
}
 
df = pd.DataFrame(data3)
 
df
```

```python
#vai retornar os valores de NaN de cada coluna  
#2 .sum retorna de todo df 

totaNaN = df.isna().sum().sum()

totaNaN 
```

```python
nanSomaA = df['A'].isna().sum()
nanSomaB = df['B'].isna().sum()
nanSomaTotal = df.isna().sum().sum()
 
 
print("+" + "-"*40 + "+")
print("| {:<15} | {:<15} |".format("Coluna", "Soma de NaN"))
print("+" + "-"*40 + "+")
print("| {:<15} | {:<15} |".format("A", nanSomaA))
print("| {:<15} | {:<15} |".format("B", nanSomaB))
print("| {:<15} | {:<15} |".format("Total Geral", nanSomaTotal))
print("+" + "-"*40 + "+")
```

```python

data6 = {
    'A': [1, 2, None, 4, 5],
    'B': [10, None, 30, 40, 50]
}
 
df = pd.DataFrame(data6)
 
df
```

```python
#preencher valores ausentes usando outros metodos específicos 

df['A'].fillna(df['A'].median(),inplace=True) # mediana no lugar dos NaN
df['B'].fillna(df['B'].ffill(),inplace=True) # preenchimento para frente 
df
```

```python
data7 = {'A': [1, 2, 3, 4, 2, 6],
         'B': [10, 20, 30, 10, 50, 60]}

df = pd.DataFrame(data7)
df
```

```python
#caça valores duplicados 1 coluna por vez 
#vai mostrar o que está duplicado e a quantidade de vezes 

duplicadas = df[df.duplicated()]
duplicadas

valoresDuplicadas = df['A'][df['A'].duplicated()].value_counts()
valoresDuplicadas
```

```python
data9 = {'A': [1, 2, 3, 4, 2, 6, 7, 2],
         'B': [10, 20, 30, 10, 50, 60, 10, 10]}

df = pd.DataFrame(data9)
df
```

```python
#$remove os duplicados por linhas se não houver ocorrencia de duplicada nas colunas A e B

removeDuplicadas = df.drop_duplicates()
removeDuplicadas
```

```python
#remoção por coluna 

removeDuplicadasA = df.drop_duplicates(subset=['A'])
removeDuplicadasA
```
